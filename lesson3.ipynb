## 巢狀判斷

```python
chinese = int(input("請輸入學生國文分數:"))

if chinese < 60:
  print("不及格!")
else:
  if chinese < 80:
    print("及格") # >= 60同時<80
  else:
    # >=60 同時 >= 80
    print("優")
print("應用程式結束")
```
