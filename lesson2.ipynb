```python
# 圓柱體體積的公式為「圓週率乘以半徑平方再乘以高」
import math

radius = 21
height = 5
radius = math.pi * radius ** 2 * height
radius
```

```python
radius = input("請輸入半徑:")
radius_int = int(radius)
height = input("請輸入高:")
height_int = int(height)

area = math.pi * radius_int ** 2 * height_int
print("圓柱體的面積是:", area)
```

```python
radius = float(input("請輸入半徑:"))
height = float(input("請輸入高:"))
area = math.pi * radius ** 2 * height
print("圓柱體的面積是:", area)
```
