````python
#學生總分為300
#有些學生可以加分5%
#如果加分超過300,就以300分為準

scores = int(input("請輸入學生分數(最高300分):"))
is_add = input("學生是否符合加分條件?(y,n)")
#print(scores,is_add)

if is_add == 'y':
  scores *= 1.05
  if scores > 300:
    scores = 300

print(f"學生分數是{round(scores)}")
````

## 如何處理錯誤

````python
try:
  money = int(input("請輸入金額:"))
  print(money)
except:
  print("輸入格式錯誤")
````

````python
#學生總分為300
#有些學生可以加分5%
#如果加分超過300,就以300分為準
try:
  scores = int(input("請輸入學生分數(最高300分):"))
  if scores > 300:
    print("分數不可以輸入大於300")
  else:
    is_add = input("學生是否符合加分條件?(y,n)")
    #print(scores,is_add)

    if is_add == 'y':
      scores *= 1.05
      if scores > 300:
        scores = 300

    print(f"學生分數是{round(scores)}")
except:
  print("出現錯誤")
````
